```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
from skimage import data
from skimage.color import rgb2gray

def add_gaussian_noise(img, mu, sigma):
    return (img + np.random.normal(mu, sigma, size=img.shape).astype(uint8) ).clip(0, 255)

def black_over_white(img):
    grayscale = 255*(1-rgb2gray(img) )
    grayscale = grayscale.astype(uint8)
    return grayscale

img = mpimg.imread('cards_db/Ravnica Allegiance/uncommon/61.jpg')
grayscale = black_over_white(img)


fig, axes = plt.subplots(1, 2, figsize=(10, 8))
ax = axes.ravel()

ax[0].imshow(img)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Black over White")

fig.tight_layout()
plt.show()
```

```python
import pytesseract

text = pytesseract.image_to_string(grayscale)

print(text)
```
